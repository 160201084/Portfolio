# Starting to build the model

#### Setting X and y

```python
X = train_a.drop(columns=['SalePrice','Id','PID'])
y = train_a['SalePrice']
```

#### Train test split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

#### We tried scaling but it did not improve our model so we will not do it

```python
# ss = StandardScaler()

# X_train_sc = ss.fit_transform(X_train)
# X_test_sc = ss.transform(X_test)
```

#### Trying out Linear Regression and checking the scores

```python
lr = LinearRegression()
lr.fit(X_train, y_train)

print(f'Linear Regression score on train: {lr.score(X_train, y_train)}')
print(f'Linear Regression score on test: {lr.score(X_test, y_test)}')
print(f'Linear Regression cross val score: {cross_val_score(lr, X_test, y_test, cv=5).mean()}')
```

#### Disabling warnings for better visualization

```python
import warnings
warnings.filterwarnings("ignore")
```

#### Ridge with scores

```python
ridge = Ridge(alpha=10)

# Fit.
ridge.fit(X_train, y_train)

# Evaluate model
print(f'Ridge score on train: {ridge.score(X_train, y_train)}')
print(f'Ridge score on test: {ridge.score(X_test, y_test)}')
print(f'Ridge cross val score: {cross_val_score(ridge, X_test, y_test, cv=5).mean()}')
```

#### RidgeCV with scores

```python
r_alphas = np.logspace(0, 5, 100)

# Cross-validate over our list of ridge alphas.
ridge_cv = RidgeCV(alphas=r_alphas, normalize=True, scoring='r2', cv=5)

# Fit model using best ridge alpha!
ridge_cv.fit(X_train, y_train);

# Evaluate model
print(f'RidgeCV score on train: {ridge_cv.score(X_train, y_train)}')
print(f'RidgeCV score on test: {ridge_cv.score(X_test, y_test)}')
print(f'RidgeCV cross val score: {cross_val_score(ridge_cv, X_test, y_test, cv=5).mean()}')
```

#### Lasso with scores

```python
l_alphas = np.logspace(-3, 0, 100)

# Cross-validate over our list of Lasso alphas.
lasso = Lasso()

# Fit model using best ridge alpha!
lasso.fit(X_train, y_train)

print(f'Lasso score on train: {lasso.score(X_train, y_train)}')
print(f'Lasso score on test: {lasso.score(X_test, y_test)}')
print(f'Lasso cross val score: {cross_val_score(lasso, X_test, y_test, cv=5).mean()}')
```

#### LassoCV

```python
l_alphas = np.logspace(-3, 0, 50)

# Cross-validate over our list of Lasso alphas.
lasso_cv = LassoCV(alphas= l_alphas, cv=5, normalize=True, max_iter=500)

# Fit model using best ridge alpha!
lasso_cv.fit(X_train, y_train)

print(f'LassoCV score on train: {lasso_cv.score(X_train, y_train)}')
print(f'LassoCV score on test: {lasso_cv.score(X_test, y_test)}')
print(f'LassoCV cross val score: {cross_val_score(lasso_cv, X_test, y_test, cv=5).mean()}')
```
